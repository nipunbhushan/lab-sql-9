```python
import pandas as pd
import numpy as np

import pymysql
from sqlalchemy import create_engine

from getpass import getpass
```

```python
password = getpass()
```

```python
connection_string = 'mysql+pymysql://root:'+password+'@localhost/sakila'
engine = create_engine(connection_string)
df_actor = pd.read_sql_query('SELECT * FROM actor', engine)
df_actor.head()
```

```python
#How many distinct (different) actors' last names are there?
len(df_actor['last_name'].unique())
```

```python
#Add an additional column day_type with values 'weekend' and 'workday' depending on the rental day of the week.
df_rental = pd.read_sql_query('SELECT * FROM rental', engine)
df_rental['day_type']=df_rental['rental_date'].apply(lambda x: 'weekday' if x.weekday() in range(0,5) else 'weekend')
```

```python
#Get all films with ARMAGEDDON in the title.
df_film = pd.read_sql_query('SELECT * FROM film', engine)
df_film[df_film['title'].str.contains("ARMAGEDDON")]
```

```python
# Get 10 the longest films.
df_film.sort_values(by='length', ascending=False).head(10)
```

```python
# How many films include Behind the Scenes content?
len(df_film[df_film['special_features'].str.contains("Behind the Scenes")])
```

```python
# Which kind of movies (rating) have a mean duration of more than two hours?
df_grp=df_film.groupby('rating').agg('mean')
df_grp[df_grp['length']>120]
```

```python
# Rank films by length (filter out the rows that have nulls or 0s in length column). In your output, only select the columns title, length, and the rank.
df_film=df_film[df_film['length'].notna() | df_film['length']!=0]
df_film['rank'] = df_film['length'].rank(method='dense',ascending=False)
df_film[['title','length','rank']].sort_values('rank')
```
